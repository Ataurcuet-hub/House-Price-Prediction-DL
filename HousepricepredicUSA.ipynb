```python
for i, j in enumerate(df['YrRenovated']):
    if j == 0:
        df.loc[i, 'house_age'] = df.loc[i, 'date'] - df.loc[i, 'YrBuilt']
    else:
        df.loc[i, 'house_age'] = df.loc[i, 'date'] - df.loc[i, 'YrRenovated']
df.head()
```
